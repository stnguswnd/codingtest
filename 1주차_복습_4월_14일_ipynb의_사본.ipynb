```python
# кІҢмһ„ л§ө мөңлӢЁ кұ°лҰ¬  https://school.programmers.co.kr/learn/courses/30/lessons/1844
from collections import deque

def solution(maps):
    cnt = 0
    row_len = len(maps)
    col_len = len(maps[0])
    visited = [[False] * col_len for _ in range(row_len)]
    visited[0][0] = True

    q = deque()
    q.append([0,0,1])

    while q:
        x,y,cnt = q.popleft()

        for dx, dy in [(-1,0),(1,0),(0,-1),(0,1)]:
            nx = x + dx
            ny = y + dy

            if nx == row_len - 1 and ny == col_len -1:
                return cnt + 1

            if nx >= 0 and nx < row_len and ny >= 0 and ny < col_len and maps[nx][ny] == 1 and visited[nx][ny] == False:
                visited[nx][ny] = True
                q.append([nx,ny,cnt+1])

    return -1
```

### 15. мң„мғҒ м •л ¬

```python
#https://www.acmicpc.net/submit/2252 мң„мғҒм •л ¬

from collections import defaultdict, deque
import sys

input = sys.stdin.readline

n, m = map(int, input().split())
answer = []

graph = defaultdict(list)
indegree = defaultdict(int)

for _ in range(m):
    a, b = map(int, input().split())
    graph[a].append(b)
    indegree[b] += 1

q = deque([i for i in range(1, n+1) if indegree[i] == 0])

while q:
    now = q.popleft()
    # лӢөм—җ м¶”к°Җ
    answer.append(now)

    # лӢӨмқҢ л…ёл“ңл“Өмқҳ м§„мһ… м°ЁмҲҳ -1
    for next_ in graph[now]:
        indegree[next_] -= 1
        if indegree[next_] == 0:
            q.append(next_)

print(*answer)
```

### 16. мҠ¬лқјмқҙл”© мңҲлҸ„мҡ°

```python
# мҠ¬лқјмқҙл”© мңҲлҸ„мҡ°/ м—°мҶҚлҗң л¶Җл¶„ мҲҳм—ҙмқҳ н•© https://school.programmers.co.kr/learn/courses/30/lessons/178870 мҳҲм ң

def solution(sequence, k):
    answer = []  # кІ°кіјлҘј лӢҙмқ„ лҰ¬мҠӨнҠё
    N = len(sequence)
    min_len = 1000000  # к°ҖмһҘ м§§мқҖ л¶Җл¶„мҲҳм—ҙмқҳ кёёмқҙлҘј м ҖмһҘн•  ліҖмҲҳ
    start = 0  # л¶Җл¶„мҲҳм—ҙмқҳ мӢңмһ‘ мқёлҚұмҠӨ
    end = 0  # л¶Җл¶„мҲҳм—ҙмқҳ лҒқ мқёлҚұмҠӨ
    window_sum = sequence[start]  # л¶Җл¶„мҲҳм—ҙмқҳ н•©мқ„ м ҖмһҘн•  ліҖмҲҳ

    while end <= N - 1:
        if window_sum == k and min_len > end - start:
            if start == end:
                return [start, end]
            min_len = end - start
            answer = [start, end]
        if window_sum < k:
            if end == N - 1:
                break
            window_sum += sequence[end + 1]
            end += 1
        else:
            window_sum -= sequence[start]
            start += 1

    return answer
```

```python
def minimize_max_workload(tasks, k):
    """
    tasks: к°Ғ нғңмҠӨнҒ¬ к°ҖмӨ‘м№ҳ лҰ¬мҠӨнҠё
    k: к°ңл°ңмһҗ мҲҳ

    л°ҳнҷҳк°’:
    к°ңл°ңмһҗл“Өм—җкІҢ нғңмҠӨнҒ¬лҘј мҲңм„ңлҘј мң м§Җн•ҳл©° лӮҳлҲҙмқ„ л•Ң,
    н•ң мӮ¬лһҢмқҙ л§ЎлҠ” к°ҖмӨ‘м№ҳ н•©мқҳ мөңлҢ“к°’мқҳ мөңмҶҢк°’
    """

    if not tasks:
        return 0

    # мөңмҶҢ к°ҖлҠҘн•ң к°’:
    # м–ҙл–Ө к°ңл°ңмһҗлҸ„ к°ҖмһҘ нҒ° нғңмҠӨнҒ¬ н•ҳлӮҳліҙлӢӨ мһ‘кІҢ л§Ўмқ„ мҲҳлҠ” м—ҶмқҢ
    left = max(tasks)

    # мөңлҢҖ к°ҖлҠҘн•ң к°’:
    # н•ң лӘ…мқҙ м „л¶Җ л§ЎлҠ” кІҪмҡ°
    right = sum(tasks)

    def can_split(max_load):
        """
        max_load мқҙн•ҳлЎң к°Ғ к°ңл°ңмһҗм—җкІҢ нғңмҠӨнҒ¬лҘј лӮҳлҲҢ мҲҳ мһҲлҠ”м§Җ нҷ•мқё
        """
        used_developers = 1
        current_sum = 0

        for weight in tasks:
            # нҳ„мһ¬ к°ңл°ңмһҗм—җкІҢ лҚ” л§Ўкёё мҲҳ мһҲмңјл©ҙ м¶”к°Җ
            if current_sum + weight <= max_load:
                current_sum += weight
            else:
                # мғҲ к°ңл°ңмһҗм—җкІҢ л„ҳк№Җ
                used_developers += 1
                current_sum = weight

                # к°ңл°ңмһҗ мҲҳ мҙҲкіј мӢң мӢӨнҢЁ
                if used_developers > k:
                    return False

        return True

    # мқҙл¶„ нғҗмғү
    while left < right:
        mid = (left + right) // 2

        if can_split(mid):
            right = mid
        else:
            left = mid + 1

    return left


# мӮ¬мҡ© мҳҲмӢң
tasks = [10, 40, 50, 10]
k = 1

answer = minimize_max_workload(tasks, k)
print("мөңмҶҢ к°ҖлҠҘн•ң мөңлҢҖ к°ҖмӨ‘м№ҳ:", answer)
```
