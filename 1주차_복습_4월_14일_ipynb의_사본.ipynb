```python
# 게임 맵 최단 거리  https://school.programmers.co.kr/learn/courses/30/lessons/1844
from collections import deque

def solution(maps):
    cnt = 0
    row_len = len(maps)
    col_len = len(maps[0])
    visited = [[False] * col_len for _ in range(row_len)]
    visited[0][0] = True

    q = deque()
    q.append([0,0,1])

    while q:
        x,y,cnt = q.popleft()

        for dx, dy in [(-1,0),(1,0),(0,-1),(0,1)]:
            nx = x + dx
            ny = y + dy

            if nx == row_len - 1 and ny == col_len -1:
                return cnt + 1

            if nx >= 0 and nx < row_len and ny >= 0 and ny < col_len and maps[nx][ny] == 1 and visited[nx][ny] == False:
                visited[nx][ny] = True
                q.append([nx,ny,cnt+1])

    return -1
```

```python
# 2 x N 타일링 https://school.programmers.co.kr/learn/courses/30/lessons/12900

#점화식만 잘 찾으면 쉬운 문제임.
def solution(n):
    dp = [0 for i in range(n)]
    dp[0], dp[1] = 1, 2
    for i in range(2, n):
        dp[i] = (dp[i-1] + dp[i-2]) % 1000000007

    return dp[n-1]
```

```python
#기지국

def solution(n, stations, w):
    answer = 0
    coverage = 2 * w + 1

    # Phase 1: [1번 아파트 옥상]부터 [마지막 기지국 커버리지 옥상]까지
    start = 1

    for station in stations:
        left, right = station - w, station + w

        # 현재 기지국의 왼쪽 범위 바깥에 전파가 미치지 않는 구간이 있다면,
        # 그 구간의 길이를 계산하고 필요한 추가 기지국의 수를 구함
        if start < left:
            length = left - start
            answer += (length + coverage - 1) // coverage # 필요한 기지국 수 계산 식

        # start 초기화
        start = right + 1

    # Phase 2: [마지막 기지국 커버리지 옥상]이 커버하지 못 하는 곳이 있다면 추가
    # start가 위에 있는 for문에서 초기화 되므로 바로 n과 비교해도 됨
    if start <= n:
        length = n - start + 1
        answer += (length + coverage - 1) // coverage

    return answer

print(solution(16,[9],2))
```

```python
# 이진 탐색 코드
def binary_search(array, target, start, end):
    if start > end:
        return None
    mid = (start+end) // 2

    if array[mid] == target:
        return mid
    elif array[mid] > target:
        return binary_search(array, target, start, mid-1)
    else:
        return binary_search(array, target, mid+1, end)

n, target = list(map(int, input().split()))

array = list(map(int, input().split()))

result = binary_search(array, target, 0, n-1)
if result == None:
    print("원소가 존재하지 않습니다.")
else:
    print(result + 1)
```

### 15. 위상 정렬

```python
#https://www.acmicpc.net/submit/2252 위상정렬

from collections import defaultdict, deque
import sys

input = sys.stdin.readline

n, m = map(int, input().split())
answer = []

graph = defaultdict(list)
indegree = defaultdict(int)

for _ in range(m):
    a, b = map(int, input().split())
    graph[a].append(b)
    indegree[b] += 1

q = deque([i for i in range(1, n+1) if indegree[i] == 0])

while q:
    now = q.popleft()
    # 답에 추가
    answer.append(now)

    # 다음 노드들의 진입 차수 -1
    for next_ in graph[now]:
        indegree[next_] -= 1
        if indegree[next_] == 0:
            q.append(next_)

print(*answer)
```

### 16. 슬라이딩 윈도우

```python
# 슬라이딩 윈도우/ 연속된 부분 수열의 합 https://school.programmers.co.kr/learn/courses/30/lessons/178870 예제

def solution(sequence, k):
    answer = []  # 결과를 담을 리스트
    N = len(sequence)
    min_len = 1000000  # 가장 짧은 부분수열의 길이를 저장할 변수
    start = 0  # 부분수열의 시작 인덱스
    end = 0  # 부분수열의 끝 인덱스
    window_sum = sequence[start]  # 부분수열의 합을 저장할 변수

    while end <= N - 1:
        if window_sum == k and min_len > end - start:
            if start == end:
                return [start, end]
            min_len = end - start
            answer = [start, end]
        if window_sum < k:
            if end == N - 1:
                break
            window_sum += sequence[end + 1]
            end += 1
        else:
            window_sum -= sequence[start]
            start += 1

    return answer
```

```python
def minimize_max_workload(tasks, k):
    """
    tasks: 각 태스크 가중치 리스트
    k: 개발자 수

    반환값:
    개발자들에게 태스크를 순서를 유지하며 나눴을 때,
    한 사람이 맡는 가중치 합의 최댓값의 최소값
    """

    if not tasks:
        return 0

    # 최소 가능한 값:
    # 어떤 개발자도 가장 큰 태스크 하나보다 작게 맡을 수는 없음
    left = max(tasks)

    # 최대 가능한 값:
    # 한 명이 전부 맡는 경우
    right = sum(tasks)

    def can_split(max_load):
        """
        max_load 이하로 각 개발자에게 태스크를 나눌 수 있는지 확인
        """
        used_developers = 1
        current_sum = 0

        for weight in tasks:
            # 현재 개발자에게 더 맡길 수 있으면 추가
            if current_sum + weight <= max_load:
                current_sum += weight
            else:
                # 새 개발자에게 넘김
                used_developers += 1
                current_sum = weight

                # 개발자 수 초과 시 실패
                if used_developers > k:
                    return False

        return True

    # 이분 탐색
    while left < right:
        mid = (left + right) // 2

        if can_split(mid):
            right = mid
        else:
            left = mid + 1

    return left


# 사용 예시
tasks = [10, 40, 50, 10]
k = 1

answer = minimize_max_workload(tasks, k)
print("최소 가능한 최대 가중치:", answer)
```
